```python
import pandas as pd
import numpy  as np
df = pd.DataFrame([\
                   ['frank', 'M',    np.nan], \
                   [np.nan , np.nan, np.nan], \
                   ['tom'  , 'M',    35], \
                   ['ted'  , 'M',    33], \
                   ['jean' , np.nan, 21], \
                   ['lisa' , 'F',    20]])
df.columns = ['name', 'gender', 'age']
df
```

```python
df.dropna()
```

```python
df.dropna(how="all")
```

```python
?df.dropna
```

```python
df.dropna(thresh=2)
```

```python
df['employee'] = np.nan
df
```

```python
df.dropna(how="all", axis = 1)
```

```python
df.fillna(0)
```

```python
df['age'].fillna(df['age'].mean())
```

```python
df['age'].fillna(df.groupby('gender')['age'].transform('mean'), inplace=True)
```

```python
df.fillna(method="pad")
```

```python
df.fillna(method="bfill", limit = 2)
```
